## Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Dataset 
(***obtained from kaggle.com***)

corona = pd.read_csv("novel-corona-virus-2019-dataset/2019_nCoV_data.csv")

```python
corona.head(10)
```

```python
corona.describe()
```
